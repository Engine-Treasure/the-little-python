```python
import pandas as pd
```

```python
# 读取 csv 到 pd.DataFrame, names 指定列标签
names1880 = pd.read_csv("ch02_material/names/yob1880.txt", names=["name", "sex", "births"])  # births 表示出生数

names1880
```

```python
names1880.groupby("sex").births.sum()  # 按性别分组, 并统计出生数
```

```python
years = range(1880, 2011)

pieces = []

columns = ['name', 'sex', 'births']
# 读取全部数据， 并为数据添加 year 列
for year in years:
    path = "ch02_material/names/yob%d.txt" % year
    frame = pd.read_csv(path, names=columns)
    frame["year"] = year  # 为某一年的数据全部加上年份信息, 新加列
    pieces.append(frame)
```

```python
pieces  # pieces is a list of DataFrame
```

```python
names = pd.concat(pieces, ignore_index=True)  # 组合全部数据

names
```

```python
# 统计出生数, 以年为行标签, 性别为列标签
total_births = names.pivot_table("births", index="year", columns="sex", aggfunc=sum)  # 按 year, sex 进行数据聚合

total_births.tail()
```

```python
%matplotlib inline

# 调用 matplotlib 绘图
total_births.plot(title="Total births by sex and year")
```

```python
def add_prop(group):
    # 类型转换, 其实 python3 的整数除法已经变了
    births = group.births.astype(float)
    # 增加新列 prop, 以婴儿数相对于总出生数的比例为值
    group["prop"] = births / births.sum()
    return group

names = names.groupby(["year", "sex"]).apply(add_prop)  # 按年份和性别分组之后, 对结果应用 add_prop 函数

names
```

```python
import numpy as np
# 合法性检查, 理论上 prop 列的总和为 1, 但浮点数无法精确到 1
# 利用近似, 只要误差足够小, 就满足合法性要求
np.allclose(names.groupby(["year", "sex"]).prop.sum(), 1)
```

```python
def get_top1000(group):
    return group.sort_values(by="births", ascending=False)[:1000]
# 按年份, 性别分组后, 再按出生函数排序, 取前 1000
grouped = names.groupby(["year", "sex"])
top1000 = grouped.apply(get_top1000)

top1000
```

```python
boys = top1000[top1000.sex == "M"]  # 布尔过滤取出男性
boys
```

```python
girls = top1000[top1000.sex == "F"]  # 布尔过滤取出女性
```

```python
# 按年份和名字的出生总数生成数据透视图
total_births = top1000.pivot_table("births", index="year", columns="name", aggfunc=sum)
```

```python
total_births
```

```python
subset = total_births[["John", "Harry", "May", "Marilyn"]]  # 取出列的子集
subset.plot(subplots=True, figsize=(12, 10), grid=True, title="Number of births per year")
```

上图反映的降低情况可能意味着父母愿意给小孩起常见名字越来越少

```python
# 1000 个最流行的名字占出生总数的比例透视图, 按年份和性别分类
table = top1000.pivot_table("prop", index="year", columns="sex", aggfunc=sum)


# np.linspace(start, end, number of parts)
table.plot(title="Sum of table1000.pop by year and sex", yticks=np.linspace(0, 1.2, 13), xticks=range(1880, 2020, 10))
```

```python
# 2010 年, 117 个最流行的男孩名字加起来才有出生总数的一半
df2010 = boys[boys.year == 2010]

prop_cumsum_2010 = df2010.sort_values(by="prop", ascending=False).prop.cumsum()

int(prop_cumsum_2010.searchsorted(0.5)) + 1  # numpy.dnarray -> int
```

```python
# 1900 年, 25个最流行的男孩名字数量加起来就达到了总数的一半
df_1900 = boys[boys.year == 1900]

in1900 = df_1900.sort_values(by="prop", ascending=False).prop.cumsum()

int(in1900.searchsorted(0.5)) + 1
```

```python
# 画出从 1880 年 到 2010 年, 所有年份的最流行名字达到出生总数 50% 的趋势
def get_quantile_count(group, q=0.5):
    group = group.sort_values(by="prop", ascending=False)
    return int(group.prop.cumsum().searchsorted(q) + 1)

diversity = top1000.groupby(["year", "sex"]).apply(get_quantile_count)
diversity = diversity.unstack("sex")

diversity.plot(title="Number of popular names in top 50%")
```

# 变成女孩名字的男孩名字(以及相反的情况)

```python
all_names = top1000.name.unique()  # 排异, 比 numpy.unique() 更快, 与 set 相比呢
all_names
```

```python
mask = np.array(["lesl" in x.lower() for x in all_names])  # 唉, 我的列表推导式啊, 用于布尔过滤

mask
```

```python
lesley_like = all_names[mask]

lesley_like
```

```python
filtered = top1000[top1000.name.isin(lesley_like)]

filtered
```

```python
filtered.groupby("name").births.sum()  # 按名字分组后, 再按出生数求和
```

```python
table = filtered.pivot_table("births", index="year", columns="sex", aggfunc="sum")  # 数据聚合

table
```

```python
table = table.div(table.sum(1), axis=0)

table.tail()
```

```python
table.plot(style={"M": "k-", "F": "k--"})
```
